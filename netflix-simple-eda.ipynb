# Exploratory Data Analysis (EDA)

```python
# visualize distribution of release_year
plt.figure(figsize=(10, 6))

plt.hist(netflix_df['release_year'], bins=20, color='blue', alpha=0.7)
plt.title('Distribution of Release Years')

plt.tight_layout()
plt.show()
```

```python
# analyze the distribution of different genres using bar plots
genre_counts = netflix_df['listed_in'].value_counts()

plt.figure(figsize=(12, 6))
genre_counts[:10].plot(kind='bar', color='purple')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Genres')
plt.ylabel('Number of Shows/Movies')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# analyze the distribution of shows/movies in different countries using bar plots
country_counts = netflix_df['country'].value_counts()

plt.figure(figsize=(12, 6))
country_counts[:10].plot(kind='bar', color='blue')
plt.title('Top 10 Movie/Show Counts per Country on Netflix')
plt.xlabel('Country')
plt.ylabel('Number of Shows/Movies')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# number of shows/movies released each year
release_year_counts = netflix_df['release_year'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.scatter(release_year_counts.index, release_year_counts.values, marker='o')
plt.title('Number of Shows/Movies Released Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Shows/Movies')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
